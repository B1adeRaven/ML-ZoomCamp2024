## Question 6

```python
median_screen = laptop_df.Screen.median()
laptop_df_with_filling = laptop_df.copy()
laptop_df_with_filling.Screen = laptop_df_with_filling.Screen.fillna(laptop_df.Screen.mode())
median_screen_with_filling = laptop_df_with_filling.Screen.median()
print(r'Median value of Screen before filling: ' + str(median_screen))
print(r'Median value of Screen after filling: ' + str(median_screen_with_filling))
print(r'Has it changed: ' + str(median_screen != median_screen_with_filling))
```
